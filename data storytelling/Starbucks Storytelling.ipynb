```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
starbucks_drink_data = pd.read_csv('starbucks_drinkMenu_expanded.csv')
```

```python
starbucks_drink_data.head()
```

```python
starbucks_drink_data=starbucks_drink_data.replace('-', np.nan)
```

```python
starbucks_drink_data=starbucks_drink_data.dropna()
```

```python
starbucks_drink_data.head()
```

```python
plt.figure(figsize = (20,5))
plt.hist(starbucks_drink_data['Calories'])
plt.xlabel('Drink Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
print(starbucks_drink_data['Beverage_category'].unique())
```

```python
#compare caffeine and calories in a scatterplot
plt.figure(figsize = (20,5))
plt.scatter(x=starbucks_drink_data['Caffeine (mg)'], y=starbucks_drink_data['Calories'], color='green')
plt.xlabel('Caffeine Content in drinks (mg)', fontsize = 15)
plt.ylabel('Calories in drinks', fontsize = 15)
plt.title('Drink Calories vs Caffeiene content', fontsize = 15)
```

```python
starbucks_drink_data['Caffeine (mg)'] = starbucks_drink_data['Caffeine (mg)'].replace(['Varies', 'varies'], np.nan)
```

```python
starbucks_drink_data['Caffeine (mg)'] = starbucks_drink_data['Caffeine (mg)'].dropna()
```

```python
starbucks_drink_data['Caffeine (mg)'].unique()
```

```python
starbucks_drink_data['Caffeine (mg)'] = pd.to_numeric(starbucks_drink_data['Caffeine (mg)'])
```

```python
#compare caffeine and calories in a scatterplot
plt.figure(figsize = (20,5))
plt.scatter(x=starbucks_drink_data['Caffeine (mg)'], y=starbucks_drink_data['Calories'], color='green')
plt.xlabel('Caffeine Content in drinks (mg)', fontsize = 15)
plt.ylabel('Calories in drinks', fontsize = 15)
plt.title('Drink Calories vs Caffeiene content', fontsize = 15)
```

```python
plt.figure(figsize = (30,5))
sns.countplot(starbucks_drink_data['Beverage_category'])
plt.xlabel('Drink Type', fontsize = 15)
plt.ylabel('Frequency', fontsize = 15)
plt.title('Beverage categories', fontsize = 15)
plt.show()
```

```python
#create a swarmplot about drink categories and calories
plt.figure(figsize=(30,10))
sns.swarmplot(data = starbucks_drink_data, x='Beverage_category', y='Calories')
plt.xlabel('Beverage Type', fontsize=15)
plt.ylabel('Calories', fontsize=15)
plt.title('Calories based on Beverage Type', fontsize=15)
plt.show()
```
